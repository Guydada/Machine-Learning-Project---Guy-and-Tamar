### Handling "nan" Data

First let's see how much is actually missing:

```python
msno.matrix(df)
plt.title('Missing Data Before Imputing', fontsize=25)
plt.show()
```

We chose to remove features of more than 50% nulls.
It is important to note as well, that the cancellation data has no missing values.

```python
# drop columns of high "null percent"
def remove_null_df(df):
    drop_col = df.isnull().sum().divide(df.shape[0])
    drop_col = list(drop_col[drop_col>0.5].index)
    df = df.drop(drop_col, axis='columns')
    return df

df = remove_null_df(df)
```

After removing those problematic values we need to examine what is left:

```python
df.isnull().sum().divide(df.shape[0]).round(8).astype(str) + "%"
```

We can observe that the missing values are a very small percent of the data. We searched for the
most common values and used it to fill the missing. 

```python
df.mode().head(1).squeeze()
```

```python
#Data imputer function

def impute_data(df):
    imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    for feature in df.columns:
        imp.fit(df[feature].values.reshape(-1,1))
        df[feature] = imp.transform(df[feature].values.reshape(-1,1))
    return df
```

```python
df = impute_data(df)
msno.matrix(df)
plt.title('Missing Data After Imputing', fontsize=30)
```

### Dealing with Categorical Data
Before turning to encoding the categorical data. We will examine two methods: LabelEncoder, and OneHotEncoder. Later we would model this 2 options and check for different results. 

```python
# Categorical Data handling

features_dropped = ['Unnamed: 0','children', 'babies',
                    'order_week', 'order_month', 'order_day_of_month',
                    'anon_feat_11', 'order_type', 'cancelation']

categorical_cols_for_encode = ['acquisition_channel',
                    'customer_type','deposit_type', 'country']

# Encoding by binary variables (OneHotEncoder)

def hot_enc(df):
    df = df.drop(features_dropped+['country_cancel_ratio'], axis='columns')
    enc = ce.OneHotEncoder(cols = categorical_cols_for_encode)
    df_bin = enc.fit_transform(df)
    return df_bin

# Encoding by lbeling (dummy variables)

def lab_enc(df):
    df = df.drop(features_dropped+['country_cancel_ratio'], axis='columns')
    enc = ce.OrdinalEncoder(cols = categorical_cols_for_encode)
    df_bin = enc.fit_transform(df)
    return df_bin

# OneHotEncoder without the country feature - using cancel ratio instead
    
def hot_enc_ratio(df):
    df = df.drop(features_dropped+['country'], axis='columns')
    enc = ce.OneHotEncoder(cols = categorical_cols_for_encode.remove('country'))
    df_bin = enc.fit_transform(df)
    return df_bin

df_hot_encoded = hot_enc(df)
df_lab_encoded = lab_enc(df)
df_hot__ratio_encoded = hot_enc_ratio(df)
```

## Dealing with outliers

```python
def outlier(X_train, y_train):
    num_train = X_train.select_dtypes(include=["number"])
    cat_train = X_train.select_dtypes(exclude=["number"])
    idx = np.all(stats.zscore(num_train) < 3, axis=1)
    X_train = pd.concat([num_train.loc[idx], cat_train.loc[idx]], axis=1)
    y_train = y_train.loc[idx]
    return X_train, y_train
```

# Part 3: Modelling <a class="anchor" id="part3"></a>

## Naive Comparison
At this stage we have the data explored and we have a few suggestions on how to make relible predictions. In order to fine tune it and put it to a test, we must compare sveral modeling estimators. <br>
We chose to run a naive test on a few of those classifiers, and later on we would hyper-tune their parameters in orer to get the best resluts. 

```python
LGR_pipe = Pipeline([('scaler', StandardScaler()),
                     ('LGR', LogisticRegression())])

KNN_pipe = Pipeline([('scaler', StandardScaler()),
                     ('KNN', KNeighborsClassifier(algorithm='brute'))])

RFC_pipe = Pipeline([('RFC', RandomForestClassifier())])

MLP_pipe = Pipeline([('MLP', MLPClassifier())])

LGR_PCA_pipe = Pipeline([('scaler', StandardScaler()),
                         ('PCA', PCA()),
                         ('LGR_PCA', LogisticRegression())])


classifiers = [
    ('LGR', LGR_pipe),
    ('KNN', KNN_pipe),
    ('RFC', RFC_pipe),
    ('MLP', MLP_pipe),
    ('LGR_PCA', LGR_PCA_pipe)
    ]
```

```python
# def Data Sets

X_datasets = {#'Hot Encoded - Model Comparing': df_hot_encoded,
              #'Label Encoded - Model Comparing': df_lab_encoded,
              'Hot Encoded - Without Country - Model Comparing': df_hot__ratio_encoded
              }
```

```python
def model_compare(X_datasets, y, classifiers, plot_scores=False):
    
    cols = [classifiers[i][0] for i in range(len(classifiers))]
    res_naive = pd.DataFrame(columns= (cols.insert(0, "Data_name")))
    i = 0
    for data_name in X_datasets:
        print("*" *100, "\nStarting test for data set {} / {}: {}".format((i+1), len(X_datasets.keys()), data_name))
        X_train, X_test, y_train, y_test = train_test_split(X_datasets[data_name], y, test_size=0.33, random_state=42)
        app_cols = {**{'Data_name': data_name}, **{col:"" for col in cols}}
        res_naive = res_naive.append(app_cols, ignore_index=True)
        for model_name, model in tqdm(classifiers):
            model.fit(X_train, y_train['cancelation'])
            kfold = model_selection.KFold(n_splits=10)
            cv_results = model_selection.cross_val_score(model, X_test, y_test['cancelation'], cv=kfold, scoring='accuracy')
            res_naive.at[i, 'Data_name'] = data_name
            res_naive.at[i, model_name] = cv_results
            msg = "%s score for %s : %f (%f)" % (type(model[model_name]).__name__, data_name, cv_results.mean(), cv_results.std())
            print(msg)
        if plot == True:
#             full_names = [type(model[name]).__name__ for name, model in classifiers]
            names = [name for name, model in classifiers]
            fig, ax = plt.subplots(figsize = (10,10))
            ax = fig.add_subplot(111)
            plt.boxplot(list(res_naive.loc[i])[1:])
            ax.set_xticklabels(names)
            plt.title('Algorithm Comparison for %s' % data_name)
            plt.show()
        i += 1
    b = res_naive.iloc[:,1:].applymap(np.mean)
    b.insert(loc = 0, column = 'Data_name', value = res_naive.iloc[:,0])
    return res_naive, b
      
```

```python
res_naive, res_naive_mean = model_compare(X_datasets, y, classifiers, plot_scores = True)
```

This df represents the way we would choose to use the data forward for each model, because as shown every conifguration currently set produced the best results.


### Hyperparameter Tuning

We are going to run the following classifiers:

```python
search_params = {
    'KNN': {'KNN__n_neighbors': list(range(1,31)),
#             'KNN__weights': ['uniform, distance'],
            'KNN__metric': ['euclidean', 'manhattan']},
    'RFC': {'RFC__n_estimators' : [100, 200, 500],
            'RFC__min_samples_split' : [2,4,6,8],
            'RFC__min_samples_leaf': [1,2,4,6],},
    'MLP': {'MLP__hidden_layer_sizes': [(50,50,50), (50,100,50), (100,)],
           'MLP__activation': ['tanh', 'relu'],
           'MLP__solver': ['sgd', 'adam'],
           'MLP__alpha': [0.0001, 0.05],
           'MLP__learning_rate': ['constant','adaptive']},
    'LGR_PCA': {'LGR_PCA__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}, 
    'LGR' : {'LGR__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
    }


def hyper_parameters(X_datasets, y, classifiers, plot=False):   
    param_res = {}
    classifiers_opt = []
    i = 0
    for data_name in X_datasets:
        print("*" *100, "\nStarting Search Grid for data set {} / {}: {}".format((i+1), len(X_datasets.keys()), data_name))
        X_train, X_test, y_train, y_test = train_test_split(X_datasets[data_name], y, test_size=0.33, random_state=42)
        for model_name, model in tqdm(classifiers):
            model.fit(X_train, y_train['cancelation'])
            param_grid = search_params[model_name]
            grid = GridSearchCV(model, param_grid=param_grid, scoring='accuracy', verbose=1, n_jobs=-1)
            grid_result = grid.fit(X_train, y_train['cancelation'])
            opt_model = model.set_params(**grid_result.best_params_) ## added to working version
            classifiers_opt.append((model_name, opt_model)) ## added to working version
            param_res[model_name] = grid_result.get_params() ## added to working version
        i+=1
    return param_res, classifiers_opt
            
```

## ROC AUC with Cross-Validation, Confusion Matrices
For the optimised models, we would like to examine the performance on the validation set using ROC AUC curves and confusion matrices

```python
def KfoldPlot(X, y, clf, clfname, k):
    
#     plt.figure(figsize=(10,10))
    
    kf = KFold(n_splits=k, shuffle=False)

    mean_tpr = np.zeros(100)
    mean_fpr = np.linspace(0, 1, 100)

    i=1
    for train_index, validation_index in tqdm(kf.split(X)):
        print("fold No. ", i)
        X_train, X_validation = X.iloc[train_index], X.iloc[validation_index]
        y_train, y_validation = y[train_index], y[validation_index]
        clf.fit(X_train, y_train)
        prob_prediction = clf.predict_proba(X_validation)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_validation, prob_prediction)
        mean_tpr += interp(mean_fpr, fpr, tpr)
        mean_tpr[0] = 0.0
        plt.plot(fpr, tpr, color='#D3D3D3')
        i+=1
    
    plt.plot([0], [0], color='#D3D3D3', linestyle='-', label='K-folds')    
    
    
    #Right now the mean_tpr is a sum (we added each iteration, so we finally divide by the number of folds)
    mean_tpr /= k
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    
    #After all of the iterations, we plot the random-guess line
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    
    #And we plot the mean roc, the red line should be in the middle of all gray lines.
    plt.plot(mean_fpr, mean_tpr, color='red', linestyle='-', label='Mean ROC (area = %0.3f)' % mean_auc)

    #Setting some boundaries and adding a legend... nothing special.
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for ' + clfname)
    plt.legend(loc="lower right")

def kfold_plot_list(X, y, classifiers, k):
    for name, model in tqdm(classifiers):
        clfname = type(model[name]).__name__
        KfoldPlot(X, y, model, clfname, k)
```

```python
# Plot confusion matrices and roc curves for every classifier on validation sets

def roc_auc_confusion(X, y, classifiers_opt, k=10):
    
    score_res = {}
    pred_res = {}
    auc_res = {}
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    for name, model in tqdm(classifiers_opt):
        fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
#         ax1, ax2 = axes.flatten()
        model.fit(X_train, y_train['cancelation'])
        score_res[name] = model.score(X_test, y_test['cancelation'])
        full_name = type(model[name]).__name__
        pred_res[name] = model.predict(X_test)
        plot_confusion_matrix(model,
                              X_test,
                              y_test['cancelation'],
                              ax=ax1, colorbar = False)
#         plot_roc_curve(model,
#                        X_train,
#                        y_train['cancelation'],
#                        ax=ax3)
#         plot_roc_curve(model,
#                        X_test,
#                        y_test['cancelation'],
#                        ax=ax4)
#         fig.delaxes(ax1)

        KfoldPlot(X, y['cancelation'], model, full_name, k=2)
        auc_res[name] = roc_auc_score(y_test['cancelation'], pred_res[name])
        print("Test AUC score for {} :\t {}\n Score res: {}".format(full_name,
                                                                    auc_res[name],
                                                                    score_res[name]))
        plt.title(full_name)
        plt.tight_layout()
        plt.show()
    return
```
